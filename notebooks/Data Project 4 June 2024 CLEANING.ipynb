```python
import pandas as pd
data = pd.read_csv('sets.csv')
```

```python
data.head()
data.info()
data.describe()
```

```python
data.dropna()
data.drop_duplicates()
```

```python
data.drop(columns=['Num_Instructions', 'Total_Quantity'], inplace=True)
```

```python
data.rename(columns={'USD_MSRP': 'Price'}, inplace=True)
data.rename(columns={'Current_Price': 'Resale_Price'}, inplace=True)
```

```python
data[['Price','Resale_Price','Pieces','Minifigures']] =  data[['Price','Resale_Price','Pieces','Minifigures']].fillna('0')
```

```python
data.fillna('{Not specified}', inplace=True)  # Replace NaN with '{Not specified}'
```

```python
data.head()
data.info()
data.describe()
```

```python
data.to_csv('sets CLEANED.csv', index=False)
```
